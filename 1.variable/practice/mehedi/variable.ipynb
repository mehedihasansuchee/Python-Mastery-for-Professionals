### 1. Integer (int)

```python
age = 20
print(age)
```

### 2. Float (float)

```python
price = 99.99
print(price)
```

### 3. String (str)

```python
name = "Hasan"
print(name)
```

### 4. Boolean (bool)

```python
is_raining = False
print(is_raining)
```

### 5. List (list)

```python
fruits = ["Apple", "Banana", "Cherry"]
print(fruits[0])
```

### 6. Tuple (tuple)

```python
coordinates = (10, 20)
print(coordinates[1])
```

### 7. Dictionary (dict)

```python
student = {"name": "Alice", "age": 22}
print(student["name"])
```

### 8. Set (set)

```python
unique_numbers = {1, 2, 3, 3, 2, 9}
print(unique_numbers)
```

```python
def say_hello():
    print("Hello, World!")
say_hello()
```

```python
def given_number(number):
    print(number * 2)

given_number(5)
given_number(10)
```

```python
def add_number(a, b):
    print(a + b)

add_number(5, 4)
add_number(9, 5)
add_number(3, 5)
```

```python
def feb_color(color):
    print("Your favorite color is" + color + "!")

feb_color("blue")
feb_color("yellow")
```

```python
def even_or_odd(number):
    if number % 2 == 0:
        print(f"{number} is even.")

    else:
        print(f"{number} is odd.")

even_or_odd(6)
even_or_odd(7)
```

```python
def multiply_numbers(numbers):
    for number in numbers:
        print(number * 5)
multiply_numbers([1, 2, 3, 4, 5, 6])
```

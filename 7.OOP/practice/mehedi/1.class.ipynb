Defining a class

```python
class Enemy:
    life = 3
    energy = 10

print(Enemy.life)
print(Enemy.energy)
```

```python
class Enemy:
    life = 5
    energy = 15

print(Enemy.life)
print(Enemy.energy)
```

class method

```python
class Enemy:
    life = 3

    def attack():
        print("life")

print(Enemy)
Enemy.attack
```

```python
class Enemy:
    life = 3
    
    @classmethod
    def attack(cls):
        print(cls.life)

print(Enemy.life)
Enemy.attack()
```
